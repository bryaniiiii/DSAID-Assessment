```python
import pandas as pd
```

```python
df = pd.read_json('https://api.covid19api.com/country/singapore/status/confirmed?from=2020-01-01T00:00:00Z&to=2022-09-01T00:00:00Z')
```

```python
df = df.groupby(df.Date.dt.to_period('Q')).nth(-1)
```

```python
df.drop(columns='Date',inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.columns.dtype
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.rc_file_defaults()
ax1 = sns.set_style(style=None, rc=None )
fig, ax1 = plt.subplots(figsize=(15,6))
plt.ticklabel_format(style='plain', axis='y')
sns.barplot(data = df, x='Date', y='Cases', alpha=0.5, ax=ax1)

for i in ax1.containers:
    ax1.bar_label(i,fmt = '%d')
    
plt.savefig('covid_cases_singapore')
```
